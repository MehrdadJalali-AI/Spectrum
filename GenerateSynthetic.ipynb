```python
import numpy as np
import matplotlib.pyplot as plt

# Set random seeds for reproducibility
np.random.seed(56)

# Function to generate a spectrum as a sum of Gaussians
def generate_spectru_gaussians(amplitudes, centers):
    widthG = 18
    x_range = np.arange(2000, 2299.8, 0.2)
    return np.sum([amplitudes[i] * np.exp(-(x_range - centers[i])**2 / widthG) for i in range(len(centers))], axis=0)

# Define specific frequencies
specificFrequencies = {
    2175: 'CeO2(110)red',
    2170: 'CeO2(110)ox',
    2176: 'CeO2(100)ox',
    2168: 'CeO2(100)red',
    2162: 'CeO2(111)red',
    2154: 'CeO2(111)ox'
}

# Generate four synthetic spectra
num_amplitudes = 6
amplitudes_list = [np.random.rand(num_amplitudes) for _ in range(4)]

# Generate spectra data
spectra_data_list = [generate_spectru_gaussians(amplitudes, list(specificFrequencies.keys())) for amplitudes in amplitudes_list]

# Define dark colors and different line styles for clear printing in both color and black-and-white
colors = ['darkblue', 'darkgreen', 'darkred', 'black']  # Dark colors for printing clarity
line_styles = ['-', '--', '-.', ':']  # Different line styles for distinction

# Plot all 4 spectra in separate subplots within one figure
fig, axs = plt.subplots(2, 2, figsize=(12, 10))  # 2 rows, 2 columns

x_range = np.arange(2000, 2299.8, 0.2)

for i, spectrum in enumerate(spectra_data_list):
    ax = axs[i // 2, i % 2]  # Position subplot
    ax.plot(x_range, spectrum, label=f"Spectrum {i+1}", linestyle='-', color=colors[i], linewidth=3)
    ax.set_xlabel("Wavenumber (CM-1)")
    ax.set_ylabel("Intensity")
    ax.invert_xaxis()  # Invert x-axis
    ax.grid(True, which="both", ls="--")


plt.tight_layout()
plt.show()



```
